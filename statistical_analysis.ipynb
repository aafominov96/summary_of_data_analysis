## Иерархический кластерный анализ

*Датасеты `beverage_r.csv` и `assess.dat.txt`*

Эта процедура предназначена для выявления относительно однородных групп наблюдений (или переменных) по заданным характеристикам при помощи алгоритма, который вначале рассматривает каждое наблюдение (переменную) как отдельный кластер, а затем последовательно объединяет кластеры, пока не останется только один.

### Алгоритм кластерного анализа:

1. Объявляется количество кластеров по количеству объектов (n).
2. Ближайшие кластеры попарно объединяются.
3. Объединение прекращается, когда расстояние между объединяемыми кластерами становится "большим".

"Большое" расстояние можно определить на основе дендрограммы:

````python
Image('5002_03.png')
````

````python
beverage = pd.read_csv("beverage_r.csv", sep=";", index_col='numb.obs')
beverage.head()
````

````python
link_beverage = sch.linkage(beverage, 'ward', 'euclidean')
````

````python
link_beverage[:5]
````

````python
dn_beverage = sch.dendrogram(link_beverage);
````

## Метод k-средних

Датасет `Protein Consumption in Europe.csv`

Метод K-средних – это метод кластерного анализа, целью которого является разделение m наблюдений на k кластеров, при этом каждое наблюдение относится к тому кластеру, к центу (центроиду) которого оно ближе всего.

### Алгоритм метода k-средних

Алгоритм разделительной кластеризации, основан на разбиении множества элементов векторного пространства на заранее определенное число кластеров k. Метод относится к неирархическим алгоритмам кластеризации. Алгоритм представляет собой итерационную процедуру:

1. Выбирается число кластеров k.
2. Из исходного множества данных случайным образом выбираются k записей, которые будут служить начальными центрами кластеров.
3. Для каждой записи исходной выборки определяется ближайший к ней центр кластера. При этом записи, «притянутые» определенным центром, образуют начальные кластеры.
4. Вычисляются центроиды – центры тяжести кластеров. Каждый центроид – это вектор, элементы которого представляют собой средние значения признаков, вычисленные по всем записям кластера. Затем центр кластера смещается в его центроид.
5. Процесс итерации прекращается, когда границы кластеров не перестанут изменяться от итерации к итерации, т.е. на каждой итерации в каждом кластере будет оставаться один и тот же набор записей.

### Пример 1

````python
protein_k_means = pd.read_csv('Protein Consumption in Europe.csv', sep=';', decimal=',', index_col='Country')
protein_k_means.head()
````

````python
norm_k_means = preprocessing.StandardScaler()
norm_k_means.fit(protein_k_means);
protein_k_means_norm = pd.DataFrame(norm_k_means.transform(protein_k_means), 
                               index=protein_k_means.index, 
                               columns=protein_k_means.columns)
protein_k_means_norm.head()
````

````python
k = range(1, 20)
models = [KMeans(n_clusters=i).fit(protein_k_means_norm) for i in k]
inert = [models[i].inertia_ for i in range(0, len(models))]
````

````python
def diff(x, y):
    d = []
    for i in range(1, len(x)):   
        d.append((y[i] - y[i-1]) / (x[i] - x[i-1]))
    return d
````

````python
plt.figure(figsize=(8, 8));
plt.plot(k, inert, marker='o', label='Sum of distances');
plt.plot(k[1:], diff(k, inert), marker='o', label='Derivative of sum')
plt.xlabel('Number of k');
plt.legend()
plt.title('The Elbow Method showing the optimal k');
plt.axvline(6, c='k', linestyle='dashed');
````

````python
protein_k_means['clusters'] = models[5].labels_
````

````python
protein_k_means.groupby('clusters').mean()
````

````python
display(pd.DataFrame(protein_k_means['clusters'].sort_values()))
````

## Проверка статистических гипотез

### Основные понятия

**Статистическая гипотеза** — предположение о виде распределения и свойствах случайной величины, которое можно подтвердить или опровергнуть применением статистических методов к данным выборки.

**Виды статистических гипотез:**
1. Гипотеза согласия - являются ли два распределения одинаковыми (часто сводится к проверке соответствия известному типу распределения)
2. Гипотеза однородности - две рассматриваемые выборки извлечены из одной и той же генеральной совокупности
3. Гипотеза независимости - две рассматриваемые случайные величины независимы
4. Гипотеза о параметре распределения - два идентичных параметра из разных выборок являются одинаковыми

**Условия применения статистических тестов:**
1. Количество наблюдение должно быть большим (как правило, должно быть больше 30).
2. Наблюдения являются случайными.
3. Вопрос должен быть сформулирован однозначно.

Как правило, решение принимается между основной(нулевой) и альтернативной гипотезой.

**Толкование решений о гипотезах:**

* **Нельзя** утверждать: "гипотеза справедлива с вероятностью XX".

* **Неточно** утверждать: "*основная гипотеза принята*"; **верно** говорить о том, что "*основная гипотеза не отвергнута*"

* **Правильно** утверждать: "*основная гипотеза отвергнута*" или "*основная гипотеза не отвергнута*"

*Пример "о неверном толковании":*

*Гипотеза: число делится на 6 нацело; инструмент для проверки: деление числа на 2. Из деления на 2 не следует делимость на 6, поэтому неверно сказать, что число делится на 6, иожно лишь не отвергнуть гипотезу.*

**Ошибки 1 и 2 рода:**
* Ошибка 1 рода заключается в том, что отвергается основная гипотеза, когда на самом деле она верна.
* Ошибка 2 рода заключается в том, что отвергается альтернативная гипотеза, когда на самом деле она верна.

*Примеры ошибок 1 и 2 рода: нулевая гипотеза - требуется срочная операция, альтернативная - операцию можно отложить. Ошибка 1 рода: срочную операцию не провели, когда нужно было её проводить, пациент умер. Ошибка 2 рода: провели срочную операцию, когда можно было не проводить, лишний раз порезали. В данной интерпретации ошибка первого рода являлась фатальной.*


````python
pd.DataFrame({'Основная гипотеза истинно верна': ['+', 'ошибка 1 рода'], 
              'Основная гипотеза истинно неверна': ['ошибка 2 рода', '+']}, 
              index=['Основная гипотеза не отвергнута', 'Основная гипотеза отвергнута'])
````

**Уровень значимости** – это вероятность того, что мы сочли различия существенными, в то время как они на самом деле случайны. Как правило, уровень значимости равен одному из трех значений: 0.05, 0.01, 0.005. Один из критериев выбоа уровня значимости - количество наблюдений (чем больше наблюдений, тем строже значение критерия).

**P-значение** — вероятность получить для данной вероятностной модели распределения значений случайной величины такое же или более экстремальное значение статистики, по сравнению с ранее наблюдаемым, при условии, что нулевая гипотеза верна.

### Проверка гипотезы о нормальности распределения случайной величины

Датасет `town_1959_2.csv`

**Основная гипотеза** - случайная величина имеет нормальное распределение, значения параметров распределения заранее не известны.

**Альтернативная гипотеза** - распределение случайной величины отличается от нормального

Саиый популярный тест проверки распределения на нормальность - **тест Шапиро-Уилка**, он предпочтительнее в использовании, если количество наблюдений меньше ~ 2000. 

В случае, если наблюдений больше, чем ~2000, то используется **тест Колмогорова-Смирнова**.

Строго говоря, допускаются некоторые отклонения от нормального распределения, при которых с распределением можно продолжать работать как с нормальным.

**Принципиальные отклонения, к которым надо относиться внимательно:**
1. Наличие выбросов в данных. К присутствию выбросов относиться **строго**. Возможный способ "лечения" - удаление выбросов.
2. Явная ассиметрия гистограммы. К отклонениям от симметрии - **снисходительно**. Возможный способ "лечения" - преобразование (например, логарифмирование или преобразование Бокса-Кокса).
3. Очень сильное отклонение гистограммы от колоколообразной формы. Отношение к колоколообразной форме гистограммы зависит от числа наблюдений. Если наблюдений меньше 30, то можно **проигнорировать** отсутствие колоколообразной формы, если от 30 до 150 наблюдений, то относиться **снисходительно**, если более 150 наблюдений, то **строго**. Возможный способ лечение, например, бимодальности - разделение на подвыборки.


````python
town_norm = pd.read_csv('town_1959_2.csv',  encoding='cp1251')
town_norm['население'].hist(bins=50);
plt.title('Гистограмма исходных данных населении');
plt.show();
np.log10(town_norm['население']).hist(bins=50);
plt.title('Гистограмма логарифмированных данных о населении');
````

````python
shapiro_test = stats.shapiro(np.log10(town_norm['население']))
shapiro_test
````

**Вывод**: нулевую гипотезу о нормальности распределение нужно отвергнуть.

### Проверка гипотезы о равенстве долей

**Основная гипотеза**: доли/вероятности разных событий равны

**Альтернативная гипотеза**: доли/вероятности разных событий различны

````python
contingency_table = pd.DataFrame([[130, 3000-130], [50, 980-50]],
                                 index=['man', 'woman'],
                                 columns=['click', 'not_click'])
display(contingency_table)
print(stats.chi2_contingency(contingency_table))
````

**Вывод**: различия в группах не являются статистически значимыми

### Сравнение центров распределений

**Центр распределения** - это одно единственное число, которое бы описывало, характеризовало выборку целиком.

В качестве центра чаще всего используют **среднее арифметическое**, **медиану** или **усеченное среднее**.

**Рекомендации по выбору центра распределения:**
1. Если распределение хотя бы одной из выборок существенно отличается от нормального, то лучше использовать **медиану**.
2. Если распределение каждой выборки можно считать нормальным или незначительно от него отличающимся, что лучше использовать **среднее арифметическое**.

**Статистические тесты** для сравнения центров распределений:
1. Если центр распределения медиана - используется **критерий Манна-Уитни-Вилкоксона** или (редко из-за высокой вероятности ошибки второго рода для небольших выборок по-сравнению с другими тестами) **критерий Mood's median**.
2. Если центр распределение среднее арифметическое - используется **одна из версий критерия Стьюдента**.


### Критерий Манна-Уитни

Датасет `Albuquerque Home Prices_data.txt`

Применять критерий Манна-Уитни для проверки гипотезы о **равенстве медиан - неверно**. Критерий проверяет насколько часто X > Y.

Критерий Манна-Уитни **не учитывает величины** наблюдений, а только лишь их **ранги**.

**Основная гипотеза**: P{X>Y} = P{X<Y} 

**Альтернативная гипотеза**: P{X>Y} != P{X<Y}

Другие названия критерия Манна-Уитни: 
* Mann-Whitney U-test
* Mann-Whitney-Wilcoxon
* Wilcoxon rank-sum test
* Wilcoxon-Mann-Whitney test

### T-критерий Стьюдента

Датасеты `agedeath.dat` и `interference.csv`

**Основная гипотеза**: математические ожидания равны

**Альтернативная гипотеза**: математические ожидания различны

* Если выборки парные, то рекомендуется использовать **парный t-критерий Стьюдента**.

* Если выборки независимые, то рекомендуется использовать **t-критерий Стьюдента для двух независимых выборое**.

**Ограничения**: 
1. Отклонения от нормального распределения должны быть несущественными (однако, стоит обратить внимание, что это требование можно проигнорировать, если наблюдений много, для точности, можно считать пороговым значением, 300 наблюдений).
2. При использовании t-критерия Стьюдента нужно сравнивать дисперсии выборок (так как существуют различные версии критериев: для одинаковых и для различных дисперсий). После проверки необходимо учитывать результат при выборе теста. Тесты для сравнения дисперсий:
   * **F-test of equality of variances** - не рекомендуется, чувствителен к отклонениям от нормальности.
   * **Bartlett's test** - лучший вариант, если данные нормально распределены. Не рекомендуется, если имеются отклонения от нормальности, так как в таком случае тест может дать ложно-положительный ответ.
   * **Levene's test**
   * **Fligner-Killeen test** - робастный, рекомендуется к использованию
   * **Brown-Forsythe test** - также возможен для применения 

#### T-Критерий Стьюдента для независимых выборок

Результаты теста говорят о том, что данные статистически значимо отклоняются от нормальных, но оценка гистограмм говорит о том, что можно принебречь результатами теста.

## A\B тесты

**A/B-тестирование** – это метод сравнения двух версий веб-страницы, электронного письма или другого маркетингового актива на основании одной переменной (одного изменяющегося элемента) с целью оптимизации эффективности его работы и улучшения важных для бизнеса показателей.

Далее будут проводиться примеры на наборе данных:

````python
belt = pd.DataFrame({'survived':[1781, 1443], 'died':[135, 47]}, index=['no seat belt', 'seat belt'])
belt
````

### Ручной Z-тест

````python
s1 = 135             #  число успехов     выборка А
n1 = 1781 + 135      #  число испытаний   выборка А
s2 = 47              #  число успехов     выборка Б
n2 = 1443 + 47       #  число испытаний   выборка Б
p1 = s1 / n1               #  оценка вероятности успеха выборка А
p2 = s2 / n2               #  оценка вероятности успеха выборка Б
p = (s1 + s2)/(n1 + n2)    #  оценка вероятности успеха выборка А+Б
z_manual = (p2 - p1) / ((p * (1 - p) * ((1 / n1) + (1 / n2))) ** 0.5)  #  z-метка
p_value_manual = stats.norm.cdf(z_manual)
print(f'z-значение при ручном расчёте: {abs(z_manual)} \np-value при ручном расчёте: {p_value_manual * 2}')
````

### Автоматический Z-тест

````python
proportion.proportions_ztest([s1, s2], [n1, n2])
````

````python
stats.chi2_contingency(belt, correction=False)
````

**Этап 4. Оценка статистической значимости полученных коэффициентов**

````python
def regression_coef(model, X, y):
    coef = pd.DataFrame({'predictor':['intercept'] + X.columns.tolist(), 
                         'coef':[model.intercept_] + model.coef_.tolist()})
    X1 = np.append(np.ones((len(X), 1)), X, axis=1)
    b = np.append(model.intercept_, model.coef_)
    MSE = np.sum((model.predict(X) - y) ** 2, axis=0) / float(X.shape[0] - X.shape[1])
    var_b = MSE * (np.linalg.inv(np.dot(X1.T, X1)).diagonal())
    sd_b = np.sqrt(var_b)
    t = b / sd_b
    coef['p_value'] = [2 * (1 - stats.t.cdf(np.abs(i), (len(X1) - 1))) for i in t]
    return coef

regression_coef(model, X, y)
````

Нетрудно заметить, что значимо отличаются от нуля лишь три предиктора: SQFT, CUST, TAX. Дополнительно стоит проверить, нужно ли включать в модель предиктор TAX, так как он является зависимым от других предикторов (тем самым, нарушается принцип коллинеарности, то есть, налог увеличивается с увеличением цены/площади). ## Полиномиальная регрессия

### Постановка задачи

Необходимо предсказывать цену кольца с бриллиантом от веса бриллианта. Попытка предсказания будет происходить не только с помощью самой переменной, но и с помощью её степеней. Первая модель будет зависеть только от самой переменной $weight$, вторая - от переменной $weight^2$, а третья - от двух этих переменных сразу.

*Ответ известен заранее, но его наличие никак не участвует в задаче: стоимость кольца с бриллиантом лучше всего предсказывает квадрат веса.*

В этих случаях, когда мы хотим попробовать в качестве фичей их степени и произведения фич, в `sklearn.preprocessing` есть класс `PolynomialFeatures`. Метод `fit_transform` этого класса сгенерирует из данного множества фич множество одночленов заданной степени. Например, для степени 2 и фич a, b будут сгенерированы фичи [a, b, a^2, b^2, ab], а при указанном параметре `include_bias=True` ещё и вектор-свободный член из единиц. Для данной задачи, конечно, проще было сгенерить один столбец из квадратов значений колонки weight. Но нам важно ознакомиться с классом `PolynomialFeatures` для дальнейшего применения.

## Прогнозирование временных рядов с помощью линейной регрессионной модели

Датасет `series_g.csv`

````python
serg = pd.read_csv('series_g.csv', sep=';')
serg['date'] = pd.to_datetime(serg['date'], format='%b %Y')
serg.head()
````

Построим графики объёма пассажироперевозок и изучим тренд и сезонность.
Виды сезонности:
1. Аддитивная - когда сезонная поправка является константой и не зависит от величины уровня тренда (иными словами, график "зажат" в между двумя кривыми, расстояние между которыми постоянно).
2. Мультипликативная - когда сезонная поправка является динамически изменяемой и зависит от величины уровня тренда (иными словами, например, когда при маленьких значениях уровня тренда сезонная поправка мала, а при больших велика)

````python
fig = plt.figure(figsize=(12, 4));
ax1 = fig.add_subplot(121);
serg['series_g'].plot(ax=ax1);
ax1.set_title('Объём пассажироперевозок');
ax1.set_ylabel('Тысяч человек');

ax2 = fig.add_subplot(122);
pd.Series(np.log10(serg['series_g'])).plot(ax=ax2);
ax2.set_title('log10 от объёма пассажироперевозок');
ax2.set_ylabel('log10 от тысяч человек');
````

Для "линеаризации" мультипликативной сезонной составляющей применяется операция логарифмирования (то есть преобразование произведения в сумму). Таким образом можно перейти к составлению линейной модели, а затем выполнить обратную операцию.
Модель будет составлена следующим образом:
$$\log y_i = \beta x_i + c(x_i) + \varepsilon_i,$$
где $y_i$ -- объём перевозок, $x_i$ -- порядковый номер месяца, $c(x_i)$ -- сезонная составляющая, $\varepsilon_i$ -- случайный шум.

Далее создадим набор данных для прогнозирования. Прогноз будет составлен для последующих 12 месяцев:

````python
serg = pd.merge(serg, pd.DataFrame({'date': pd.date_range('1949-01-01', '1961-12-01', freq='MS')}), on='date', how='right')
````

Создадим регрессионную переменную ```month_num``` - порядковый номер пары (месяц, год), а также прологарифмируем целевой показатель.

````python
serg['month_num'] = range(1, len(serg) + 1)
serg['log_y'] = np.log10(serg.series_g)
````

Далее, для учета фактора сезонности создадим 12 колонок (по количеству месяцев).

````python
for x in range(1, 13):
    serg['season_' + str(x)] = serg['date'].dt.month == x
````

При таком подходе к учёту сезонности нарушается линейная независимость переменных, поэтому необходимо исключить из рассмотрения один из месяцев сезонности. Пусть базовым месяцем будет январь (формально, мы преобразовали набор таким образом, что январь исключен, а остальные месяцы учтены с поправкой на январь).

Создадим обучающие векторы:

````python
season_columns = ['season_' + str(x) for x in range(2, 13)]
X = serg[['month_num'] + season_columns]
X1 = X[X.index < 144]
y = serg['log_y']
y1 = y[y.index < 144]
````

Перейдем к настройке модели:

````python
model_serg = LinearRegression()
model_serg.fit(X1, y1);
````

Проверим, как модель отработала обучающую выборку:

````python
pd.DataFrame({'predict': model_serg.predict(X1), 'real': y1}).plot(figsize=(12, 6));
````

Значение $R^2$:

````python
model_serg.score(X1, y1)
````

Теперь построим предсказание для всей матрицы, включая неизвестные месяцы, а также провеедем обратное преобразование данных из логарифмического масшатаба в реальный:

````python
pd.DataFrame({'predict': 10 ** model_serg.predict(X), 'real': 10 ** y}).plot(figsize=(12, 6));
````

## Деревья решений

**Дерево решений** — метод автоматического анализа больших массивов данных.

**Дерево решений** — эффективный инструмент интеллектуального анализа данных и предсказательной аналитики. Он помогает в решении задач по классификации и регрессии.

Дерево решений представляет собой иерархическую древовидную структуру, состоящую из правила вида «Если …, то ...». За счет обучающего множества правила генерируются автоматически в процессе обучения.

В отличие от нейронных сетей, деревья как аналитические модели проще, потому что правила генерируются на естественном языке: например, «Если реклама привела 1000 клиентов, то она настроена хорошо».

Правила генерируются за счет обобщения множества отдельных наблюдений (обучающих примеров), описывающих предметную область. Поэтому их называют индуктивными правилами, а сам процесс обучения — индукцией деревьев решений.

В обучающем множестве для примеров должно быть задано целевое значение, так как деревья решений — модели, создаваемые на основе обучения с учителем. По типу переменной выделяют два типа деревьев:

* дерево классификации — когда целевая переменная дискретная;

* дерево регрессии — когда целевая переменная непрерывная.

### Деревья решений в задачах классификации

````python
credit = pd.read_csv('Credit.csv', sep=';', encoding='cp1251')
credit.head()
````

````python
# Правильный ответ записываем в вектор y
y_credit = credit['кредит']
# Удаляем колонку с правильным ответом
X_credit = credit.drop('кредит', axis=1)
````

````python
model_credit = DecisionTreeClassifier(random_state=42,
                                      # функция для impurity ('gini' или 'entropy')
                                      criterion='gini',
                                      # максимальная глубина дерева
                                      max_depth=5,
                                      # минимальное число элементов в узле для разбиения (может быть долей)
                                      min_samples_split=5,
                                      # минимальное число элементов в листе (может быть долей)
                                      min_samples_leaf=5,
                                      # веса для классов (можно дополнительно штрафовать за ошибку в нужных классах).
                                      # поддерживает опцию 'balanced'.
                                      class_weight=None,
                                      )
model_credit.fit(X_credit, y_credit);
````

Для интерпретации получившийся модели удобно отобразить её в виде дерева предикатов (решающих правил). Для этого можно воспользоваться функцией `export_graphviz` библиотеки `sklearn`.

````python
export_graphviz(model_credit,
               out_file='tree.dot',
               class_names=None,
               label='all',
               filled=True,
               impurity=True,
               node_ids=True,
               proportion=True,
               rotate=True)
````

````python
render('dot', 'png', 'tree.dot')
Source.from_file("tree.dot")
````

Модель позволяет оценить ценность каждой фичи в смысле её эффективности для разбиения (Gini importance). Эта оценка считает для каждой из фич сумму дельт по impurity (нормированную), полученных при разбиениях по этой фиче. Для этого у модели есть метод `feature_importances_`.

````python
pd.DataFrame({'Feature': X_credit.columns, 
              'Importance': model_credit.feature_importances_}).sort_values(by='Importance', ascending=False)
````
